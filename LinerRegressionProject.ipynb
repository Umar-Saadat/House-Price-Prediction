```python
import pandas as pd

from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score

import matplotlib.pyplot as plt

import warnings

from sklearn.model_selection import train_test_split

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('house.csv')

print(data.head())
```

```python
plt.scatter(data['size_sqft'],data['price'])
plt.show()
```

```python
features = data[['size_sqft','bedrooms','age_years']]

Prices = data['price']
```

```python
model = LinearRegression()

model.fit(features,Prices)
```

```python
size = float(input('Enter the size in sqft : '))

bedrooms = float(input('Enter the Bedrooms : '))

age = float(input("Enter the age of home : "))

prediction_price = model.predict([[size,bedrooms,age]]).item()


print(f"Predicted Price is : {prediction_price:.2f}")
```
